```python
import numpy as np
import cv2
import os
from keras.models import Sequential
from keras.layers import Activation, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import load_img, img_to_array, array_to_img
```

```python
INPUT_WIDTH = 100
INPUT_HEIGHT = 100
BATCH_SIZE = 128
EPOCHS = 100
STEPS = 5
CATEGORY = 4

OK_DIR1 = "C:\\Users\\kaida\\deeplearning\\car\\data\\ok_body_only"
OK_DIR2 = "C:\\Users\\kaida\\deeplearning\\car\\data\\ok_other"
NG_DIR1 = "C:\\Users\\kaida\\deeplearning\\car\\data\\ng_kizu"
NG_DIR2 = "C:\\Users\\kaida\\deeplearning\\car\\data\\ng_hekomi"
```

```python
def load_image(dir_name, label):
    x = []
    y = []
    files = os.listdir(dir_name)
    for file in files:
        img = cv2.imread(os.path.join(dir_name, file))
#        img = cv2.imread(os.path.join(dir_name, file), 0)
        # リサイズ
        img = cv2.resize(img, (INPUT_WIDTH, INPUT_HEIGHT))
        # 正規化
#        img = (img - np.mean(img)) / np.std(img)

        img_array = np.array(img_to_array(img))
        x.append(img_array)
        if label == 0:
            y.append([1, 0, 0, 0]) # OK1ラベル
        elif label == 1:
            y.append([0, 1, 0, 0]) # OK1ラベル
        elif label == 2:
            y.append([0, 0, 1, 0]) # NG1ラベル
        else:
            y.append([0, 0, 0, 1]) # NG2ラベル
    
    return (x, y)
```

```python
x_train = []
y_train = []
x_test = []
y_test = []

# OK画像を読み込む
x_ok1, y_ok1 = load_image(OK_DIR1, 0)
x_ok2, y_ok2 = load_image(OK_DIR2, 1)
x_ng1, y_ng1 = load_image(NG_DIR1, 2)
x_ng2, y_ng2 = load_image(NG_DIR2, 3)

x_train = x_ok1 + x_ok2 + x_ng1 + x_ng2
y_train = y_ok1 + y_ok2 + y_ng1 + y_ng2
    
# np配列化
x_train = np.array(x_train)
y_train = np.array(y_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for i in range(10):
    plt.subplot(4, 5, i+1)
    plt.imshow(array_to_img(x_train[i]))
    plt.gray()
plt.show() 
```

```python
# 訓練データとテストデータを分割
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.3, random_state=None)
```

```python
# モデルを作る
def cnn_model():
    model = Sequential()
    
    # 畳み込み層１
    model.add(Conv2D(32, (3, 3), padding="SAME", input_shape=[INPUT_WIDTH, INPUT_HEIGHT, 3], name="conv1"))
    model.add(Activation("relu"))
    model.add(MaxPooling2D(pool_size=(2, 2))) # -> 50 x 50
    
    # 畳み込み層２
    model.add(Conv2D(32, (3, 3), padding="SAME", name="conv2"))
    model.add(Activation("relu"))
    model.add(MaxPooling2D(pool_size=(2, 2))) # -> 25 x 25
    
    # 畳み込み層３
    model.add(Conv2D(64, (3, 3), padding="SAME", name="conv3"))
    model.add(Activation("relu"))
    model.add(MaxPooling2D(pool_size=(2, 2))) # -> 12 x 12
    model.add(Dropout(0.8))
    
    model.add(Flatten())
    # 全結合１
    model.add(Dense(64, name="dense1"))
    model.add(Activation("relu"))
    model.add(Dropout(0.5))

    # 全結合２
#    model.add(Dense(64, name="dense2"))
#    model.add(Activation("relu"))
#    model.add(Dropout(0.5))
    
    # 出力
    model.add(Dense(CATEGORY, name="output"))
    model.add(Activation("softmax"))
    
    return model
```

```python
model = cnn_model()
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
train_gen = ImageDataGenerator(
#    rescale=1./255,
#    featurewise_center=False,
#    samplewise_center=False,
#    featurewise_std_normalization=False,
#    samplewise_std_normalization=False,
    rotation_range=180,
#    width_shift_range=0.2,
#    height_shift_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
#    zoom_range=[.8, 1],
#    channel_shift_range=30,
#    fill_mode='reflect'
)
test_gen = ImageDataGenerator()
```

```python
# kerasは0～1しか受け付けない
x_train = x_train / 255
y_train = y_train / 255
```

```python
history = model.fit_generator(train_gen.flow(x_train, 
                                  y_train, 
                                  batch_size=BATCH_SIZE, 
                                  shuffle=True
                                 ), 
                    steps_per_epoch=STEPS, 
                    epochs=EPOCHS,
                    #callbacks=cbks,
                    validation_data=(x_test, y_test))
```

```python
def predict_message(predict):
    indice = np.argmax(predict)
    if indice in {0, 1}:
        print("ok (%s)" % str(predict[0] + predict[1]))
    elif indice in {2}:
        print("ng kizu (%s)" % str(predict[indice]))
    elif indice in {3}:
        print("ng hekomi (%s)" % str(predict[indice]))
```

```python
def predict_ok_ng(x, y):
    correct = 0
    wrong = 0
    predicts = model.predict(x)
    for i in range(len(x)):
        indice = np.argmax(predicts[i])
        # ok
        if y[i][0] == 1 or y[i][1]:
            if indice in {0, 1}:
                correct += 1
            else:
                wrong += 1
        else:
            if indice in {2, 3}:
                correct += 1
            else:
                wrong += 1
    
    print("correct: %s / wrong: %s / acc: %s" % (str(correct), str(wrong), str(correct / (correct + wrong))))
```

```python
predict_ok_ng(x_test, y_test)
```
